```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Read the CSV file from Google Drive
df = pd.read_csv('/content/drive/My Drive/sex_ed_data.csv')  # adjust path if needed

# Preprocess
df = df[['Question', 'Grade']].dropna()

# Feature and target
X = df['Question']
y = df['Grade']

# Convert text to vectors
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# Train the model
model = LogisticRegression()
model.fit(X_vec, y)
```

```python
def predict_grade(question):
    vec = vectorizer.transform([question])
    return model.predict(vec)[0]
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "google/gemma-2b-it"  # You can also try "google/gemma-1.1-2b-it"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

text_gen = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
def generate_answer_with_gemma(question, grade):
    prompt = f"Answer the following sex education question in an age-appropriate way for a student in Grade {grade}:\n\n{question}"
    output = text_gen(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)[0]['generated_text']
    return output[len(prompt):].strip()

question = "What is masturbation?"
grade = 8

print(generate_answer_with_gemma(question, grade))
```
